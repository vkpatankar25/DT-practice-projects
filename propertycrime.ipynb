# Check the State wise prercentage of crime

```python
df['Difference']=(df["Cases_Property_Stolen"]- df["Cases_Property_Recovered"])/df['Cases_Property_Stolen']*100
print(df[['Difference','Area_Name']])
```

```python
df['Difference1']=(df["Value_of_Property_Stolen"]- df["Value_of_Property_Recovered"])/df['Value_of_Property_Stolen']*100
print(df[['Difference1','Area_Name']])
```

```python
df['Difference'].sort_values().mean()
```

```python
df['Difference1'].sort_values().mean()
```
